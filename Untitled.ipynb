```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
df.head(10)
```

```python
def divisible_by_ten(nums):
    if (nums % 10 == 0):
      return True
    else:
        return False
    
assert divisible_by_ten(20)==False
```
